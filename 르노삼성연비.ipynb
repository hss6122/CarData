# 결측치 있는 행 제거

# 오브젝트 타입을 int 형식으로 바꿔줘야할 형식이 있음
# 배기량,공차중량,도심주행연비,고속도로주행연비,복합연비

```python
df.describe() # 각 항목을 정리해주는데. 정수,실수 타입만 됨... 흑
```

```python
df['배기량'] = pd.to_numeric(df['배기량'],errors='coerce')
df['공차중량'] = pd.to_numeric(df['공차중량'],errors='coerce')
df['도심주행연비'] = pd.to_numeric(df['도심주행연비'],errors='coerce')
df['고속도로주행연비'] = pd.to_numeric(df['고속도로주행연비'],errors='coerce')
df['복합연비'] = pd.to_numeric(df['복합연비'],errors='coerce')
```

```python
df.info()
```

```python
df.groupby('유종').size()
```

```python
df휘발유 = df.groupby('유종').get_group('휘발유')
```

```python
df휘발유
```

```python
plt.figure(figsize=[17,5])

plt.subplot(131) # 인덱스 1번 그래프
sns.barplot('고속도로주행연비', '모델명', data=df휘발유) 

plt.subplot(132) # 2번 그래프
sns.barplot('도심주행연비', '모델명', data=df휘발유) #나이가 적을 수록 생존률이 높다

plt.subplot(133)
sns.barplot('복합연비', '모델명', data=df휘발유)

plt.subplots_adjust(top=1, bottom=0.1, left=0.10, right=1, hspace=0.5, wspace=0.5)

plt.show()
```

```python
sns.pairplot(df휘발유, hue='모델명', vars=['고속도로주행연비','도심주행연비','복합연비'])
```

```python
plt.scatter(df휘발유.도심주행연비, # x
           df휘발유.모델명, # y
           alpha=0.2,
            s=20*df휘발유.도심주행연비,
            c=df휘발유.도심주행연비, # color
           cmap='viridis'
           ) # x
plt.title('르노삼성(휘발류)차량 도심주행연비', fontsize=14)
plt.xlabel('도심주행연비 km/ℓ',fontsize=12)
plt.ylabel('모델명',fontsize=12)
plt.colorbar()
plt.show()
```

```python
plt.scatter(df휘발유.고속도로주행연비, # x
           df휘발유.모델명, # y
           alpha=0.2,
            s=20*df휘발유.고속도로주행연비,
            c=df휘발유.고속도로주행연비, # color
           cmap='viridis'
           ) # x
plt.title('르노삼성(휘발류)차량 고속도로주행연비', fontsize=14)
plt.xlabel('고속도로주행연비 km/ℓ',fontsize=12)
plt.ylabel('모델명',fontsize=12)
plt.colorbar()
plt.show()
```

```python
plt.scatter(df휘발유.복합연비, # x
           df휘발유.모델명, # y
           alpha=0.2,
            markers=["o", "x", "+", "*"],
            s=20*df휘발유.복합연비,
            c=df휘발유.복합연비, # color
           cmap='viridis'
           ) # x
plt.title('르노삼성(휘발류)차량 복합연비', fontsize=14)
plt.xlabel('복합연비 km/ℓ',fontsize=12)
plt.ylabel('모델명',fontsize=12)
plt.colorbar()
plt.show()
```
